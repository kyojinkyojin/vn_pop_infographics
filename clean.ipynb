```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("./V02.01.xlsx")
df.columns = df.iloc[1]
df = df.iloc[3:]
df = df.dropna(how='all').reset_index(drop=True)
df.columns = ['Region', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

non_province = ["CẢ NƯỚC", "Đồng bằng sông Hồng", "Trung du và miền núi phía Bắc", "Bắc Trung Bộ và Duyên hải miền Trung", "Đông Nam Bộ", "Đồng bằng sông Cửu Long"]
province = [r for r in df["Region"] if r not in non_province]

df = df[df["Region"].isin(province)].reset_index(drop=True)
df['Region'] = df['Region'].str.replace("oà", "òa")
df['Region'] = df['Region'].str.replace("oá", "óa")
col_to_convert = [str(year) for year in range(2011, 2023)]
df[col_to_convert] = df[col_to_convert].astype(str)

def add_prefix(cn):
    try: 
        year = int(cn)
        if 2011 <= year <= 2022:
            return f"y_{year}"
    except ValueError:
        pass
    return cn

df = df.rename(columns=add_prefix)


df.info()
```

```python
vndf = pd.read_csv("./Vietnam_properties_f.csv", sep=";", encoding="UTF-8")
merged_df = pd.merge(vndf, df, how='left', left_on='name', right_on='Region')
merged_df = merged_df.drop(columns=['Region'])
vndf.head()
```

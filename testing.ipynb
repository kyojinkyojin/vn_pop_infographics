```python
import pandas as pd
import numpy as np

excel_raw = pd.ExcelFile("./raw/indepth_pop.xlsx")
df = pd.read_excel(excel_raw, "Bieu 1")
```

```python
df.columns = df.iloc[3]
df.columns = ["Region", "all_total", "all_m", "all_f", "city_total", "city_m", "city_f", "rural_total", "rural_m", "rural_f"]
df = df.iloc[4:].reset_index(drop=True)
df.head()
```

```python
df = df.drop(1)
df.head()
```

```python
non_province = ["CẢ NƯỚC", "Đồng bằng sông Hồng", "Trung du và miền núi phía Bắc", "Bắc Trung Bộ và Duyên hải miền Trung", "Đông Nam Bộ", "Đồng bằng sông Cửu Long", "TOÀN QUỐC", "Tỉnh, thành phố"]
province = [r for r in df["Region"] if r not in non_province]

df = df[df["Region"].isin(province)].reset_index(drop=True)
df['Region'] = df['Region'].str.replace("oà", "òa")
df['Region'] = df['Region'].str.replace("oá", "óa")
df.head()
```
